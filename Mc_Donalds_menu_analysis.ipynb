```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("menu.csv")
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 4), dpi=100) #Input plot size
menu_category = df.Category.value_counts()
menu_category.plot.bar(color = ['red','green','green','green','green','green','green','green','green']) #graph color
plt.title("Number of Menu Items for each Food Catagory")
plt.ylabel("Number of Item's")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

correlation = df['Calories'].corr(df['Calories from Fat'])
plt.scatter(df.Calories, df['Calories from Fat'], color='blue')
plt.text(500,450, 'r = {}'.format(round(correlation,2)))
plt.xlabel("Calories")
plt.ylabel("Calories form Fat")
plt.title("The Relationship Between Calories vs Calories from Fat")
plt.show()
```

```python
#Bar Graph for Percent of Trans Fat
plt.figure (figsize=(10, 4), dpi=100) #input plot size
menu_category = df.groupby('Category')['Trans Fat'].mean()
menu_category.plot.bar(color = ['blue','green','green','green','green','green','green','green','green'])
plt.title("Percent of Trans Fat")
plt.xlabel("Trans Fat")
plt.ylabel("Categories")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby('Category')['Trans Fat'].mean() # Trans Fat Categories
```

```python
df['saturated_cholesterol'] = df['Saturated Fat']/df['Cholesterol']*100
```

```python

saturated_cholesterol = df.groupby('Category')['saturated_cholesterol'].mean().dropna().nlargest(5)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
saturated_cholesterol.sort_values(ascending=False).plot.bar(color = 'violet')
plt.title("Top 5 Categories that have the Highest Percent of Saturated Fat")
plt.ylabel("Percent")
plt.xlabel("Menu Category")
plt.xticks(rotation=90)
plt.show();
```

```python
coffee_tea = df[df.Category == 'Coffee & Tea']
coffee_tea.groupby('Item')['saturated_cholesterol'].mean().sort_values(ascending=False)
```

```python
shakes = df[df.Category == 'Smoothies & Shakes']
shakes.groupby('Item')['saturated_cholesterol'].mean().sort_values()
```

```python
beef_pork = df[df.Category == 'Beef & Pork']
beef_pork.groupby('Item')['saturated_cholesterol'].mean().sort_values()
```

```python
chicken_fish = df[df.Category == 'Chicken & Fish']
chicken_fish.groupby('Item')['saturated_cholesterol'].mean().sort_values()
```

```python
df.groupby('Category')['Vitamin A (% Daily Value)'].mean()
```

```python
df.groupby('Category')['Vitamin C (% Daily Value)'].mean()
```

```python
beverage = df[df.Category == 'Beverages']
beverage.groupby('Item')['Vitamin C (% Daily Value)'].mean().sort_values()
```

```python
iron = df.groupby('Category')['Iron (% Daily Value)'].mean()
calcium = df.groupby('Category')['Calcium (% Daily Value)'].mean()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

category = iron.index
Iron = iron.values
Calcium = calcium.values
X_axis = np.arange(len(category))

plt.bar(X_axis - 0.2, Iron, 0.4, label = 'Iron', color = 'blue')
plt.bar(X_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'green')

plt.xticks(X_axis, category)
plt.xlabel("Menu Categories")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
